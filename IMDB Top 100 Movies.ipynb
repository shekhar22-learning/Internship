```python
from bs4 import BeautifulSoup
import requests
```

```python
source=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
```

```python
print(source)
```

```python
imdbSoup=BeautifulSoup(source.content)
```

```python
imdbSoup
```

```python
movies = imdbSoup.find('tbody', class_="lister-list").find_all('tr')
```

```python
movies
```

```python
sheet=[]

for movie in movies:
    name = movie.find('td', class_="titleColumn").a.text
    year = movie.find('td', class_="titleColumn").span.text.strip('()')
    rating = movie.find('td', class_="ratingColumn imdbRating").strong.text
    
    sheet.append({'Movie Name': name,'Year of Release' : year,'IMDB Rating': rating})
    
    print(name,year,rating)
    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(sheet)
```

```python
df
```
