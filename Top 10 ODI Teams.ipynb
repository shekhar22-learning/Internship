```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
icc_soup=BeautifulSoup(page.text)
```

```python
icc_soup
```

```python
team = []

for i in icc_soup.find_all('span',class_="u-show-phablet"):
    team.append(i.text)
```

```python
match = []

for i in icc_soup.find_all('td', class_="table-body__cell u-center-text"):
    match.append(i.text)
```

### The above code is Extracting values of both 'Matches' and 'Points'

```python
match
```

#### So, here I am segregating the data from "match" by creating 2 DataFrames "matches" and "points" 

#### Besides that, it is also not extracting the values of 'First Row' from the website

```python
df = pd.DataFrame(match)
matches = df.iloc[::2]
```

```python
print(matches)
```

#### So, I am adding 'First Row Value' manually for country "NZ"

```python
new_row_matches = pd.DataFrame({17},index=[0])
matches=pd.concat([new_row_matches, matches]).reset_index(drop=True)
```
