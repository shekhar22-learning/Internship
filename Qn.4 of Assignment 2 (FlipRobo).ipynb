```python
# Importing the required libraries

import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r"F:\SOFTWARE\Web Drivers\chromedriver_win32\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
url = "https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
driver.get(url)
```

```python
#Creating Empty List for the Data to be Extracted

brand_list = []
prod_desc_list = []
price_list = []
discount_list = []
```

```python
brand_tags = driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
brand_tags[0:5]
```

```python
for i in brand_tags:
    brand=i.text
    brand_list.append(brand)
brand_list[0:5]
```

```python
prod_tags = driver.find_elements_by_xpath("//a[@class='IRpwTa']")
prod_tags[0:5]
```

```python
for i in prod_tags:
    prod=i.text
    prod_desc_list.append(prod)
prod_desc_list[0:5]
```

```python
price_tags = driver.find_elements_by_xpath("//div[@class='_30jeq3']")
price_tags[0:5]
```

```python
for i in price_tags:
    price=i.text
    price_list.append(price)
price_list[0:5]
```

```python
disc_tags = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
disc_tags[0:5]
```

```python
for i in disc_tags:
    disc=i.text
    discount_list.append(disc)
discount_list[0:5]
```

```python
print(len(brand_list), len(prod_desc_list), len(price_list), len(discount_list))
```

```python
page1 = pd.DataFrame({})
page1['Brand']=brand_list[0:38]
page1['Product Desc']=prod_desc_list[0:38]
page1['Price']=price_list[0:38]
page1['Discount']=discount_list[0:38]
```

```python
page1
```
